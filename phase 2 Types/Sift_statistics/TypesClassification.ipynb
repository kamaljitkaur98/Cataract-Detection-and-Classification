```python
#Basic Imports
import cv2
import os
import numpy as np
```

```python
categories = ['healthy eyes','mild cataract', 'severe cataract']
directory = r'C:\Users\kamal\ishitaProject\Dataset'
dataset = []
for category in categories:
    path = os.path.join(directory,category)
    label = categories.index(category)
    for image in os.listdir(path):
        image_path = os.path.join(path,image)
        data_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        data = cv2.GaussianBlur(data_image,(5,5),cv2.BORDER_DEFAULT)
        mean,std = cv2.meanStdDev(data)
        dataset.append([mean[0][0],std[0][0],label])
print("Mean \t std \t label")
mean_healthy = []
mean_severe = []
mean_mild = []
for mean,std,label in dataset:
    category = ""
    if(label == 0):
        category = "Healthy"
        mean_healthy.append(mean)
    elif(label == 1):
        category = "Mild"
        mean_mild.append(mean)
    else:
        category = "Severe"
        mean_severe.append(mean)
    print("{:.2f} \t {:.2f} \t {} ".format(mean,std,category))
```

```python
#Plotting comparative histogram
from matplotlib import pyplot as plt
```

```python
plt.hist(mean_healthy, bins = [0,50,100,125,150],color='green')
plt.show()
```

```python
plt.hist(mean_mild, bins = [0,50,100,125,150], color='blue')
plt.show()
```

```python
plt.hist(mean_severe, bins = [0,50,100,125,150], color='red')
plt.show()
```

```python
import os
subdirs = [x[0] for x in os.walk('dataset')]
print(subdirs)
```

```python
from os import listdir
from os.path import isfile, join
file_names=[]
for sub_dir in subdirs:
    if sub_dir == 'dataset\\normal_png':
        continue
    mypath = sub_dir
    sub_folder_file_names = [mypath+"/"+f for f in listdir(mypath) if isfile(join(mypath, f))]
    file_names += sub_folder_file_names
    
print(str(len(file_names)) + ' images loaded')
```

```python
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
size=100
hei=[]
wei=[]
c=0
for filename in file_names:
    img=cv2.imread(filename)
    img = cv2.resize(img, (size, size), interpolation = cv2.INTER_AREA)
    oimg=img.copy()
    kernel=np.ones((3,3),np.uint8)
    cimg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    #80 30
    clahe = cv2.createCLAHE(clipLimit = 5) 
    final_img = clahe.apply(cimg) + 30
    # 80 33 for clahe size 80
    #80 42 for 100
    circles = cv2.HoughCircles(final_img,cv2.HOUGH_GRADIENT,1,20,param1=80,param2=42,minRadius=0,maxRadius=0)
    if circles is not None:
        c+=1
        detect_circles = np.uint16(np.around(circles)) 
        max_a=0
        for i in detect_circles[0,:]:
            # draw the outer circle
            ar=int(i[2]*i[2]*math.pi)
            if ar>max_a:
                x=i[0]
                y=i[1]
                r=i[2]
                max_a=ar
        #cv2.circle(oimg,(x,y),r,(0,255,0),2)
        #cv2.imshow(filename,oimg)
        #cv2.waitKey(0)
        #cv2.destroyAllWindows()
        x=int(x)
        y=int(x)
        r=int(x)
        height,width=cimg.shape
        mask = np.zeros((height,width), np.uint8)
        cv2.circle(mask,(x,y),r,(255,255,255),-1)
        masked_data = cv2.bitwise_and(img, img, mask=mask)
        _,thresh = cv2.threshold(mask,1,255,cv2.THRESH_BINARY)

        # Find Contour
        cnt = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[0]

        #print len(contours)
        x,y,w,h = cv2.boundingRect(cnt[0])

        # Crop masked_data
        crop = masked_data[y:y+h,x:x+w]
        mild_path="preprocessed_40\mild/"
        severe_path="preprocessed_40\severe/"
        normal_path="preprocessed_40/normal/"
        if filename[8] == 'm':        
            print(cv2.imwrite( mild_path+filename[13:] , crop ))
        elif filename[8] == 's':     
            print(cv2.imwrite( severe_path+filename[15:] , crop ))
        else:
            print(cv2.imwrite( normal_path+filename[15:] , crop ))
print(c)
```
